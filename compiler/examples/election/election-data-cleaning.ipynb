```python
import pandas as pd
```

```python
state_csv = pd.read_csv("usmap-state.csv")
state_csv.head(10)
```

```python
import random
# shuffled_states = random.shuffle()
state_csv = state_csv.sample(frac=1).reset_index(drop=True)
state_csv.head(10)
```

```python
# print(state_csv.at[1, 'crimerate'])
state_csv = state_csv[state_csv.name.notnull()]
for idx in range(len(state_csv)):
    state_csv.at[idx, 'crimerate'] = random.random() * 100.0
state_csv.rename(columns={'crimerate':'dem_voteshare'}, inplace=True)
state_csv.head(10)
```

```python
county_csv = pd.read_csv("usmap-county.csv")
county_csv.head(10)
```

```python
county_csv = county_csv.sample(frac=1).reset_index(drop=True)
county_csv = county_csv[county_csv.name.notnull()]

for idx in range(len(county_csv)):
    county_csv.at[idx, 'crimerate'] = random.random() * 100.0
    
county_csv.rename(columns={'crimerate':'dem_voteshare'}, inplace=True)
county_csv.head(10)
```

```python
state_csv.to_csv("election-states.csv")
county_csv.to_csv("election-counties.csv")
```
