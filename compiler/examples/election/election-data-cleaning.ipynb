```python
import pandas as pd
import random
```

```python
county_csv = pd.read_csv("usmap-county.csv")
county_csv.head(10)
```

```python
#county_csv = county_csv.sample(frac=1).reset_index(drop=True)
county_csv = county_csv[county_csv.name.notnull()]
county_csv.rename(columns={'population':'total_votes'}, inplace=True)
del county_csv['crimerate']

print("len county csv: ", len(county_csv))
dem_votes_col = []
rep_votes_col = []
for idx in range(len(county_csv)):
    dem_votes = int(random.random() * county_csv.at[idx, 'total_votes'])
    dem_votes_col.append(dem_votes)
    rep_votes = int(county_csv.at[idx, 'total_votes'] - dem_votes)
    rep_votes_col.append(rep_votes)

county_csv['state_id'] = [int(sid) for sid in county_csv['state_id']]
county_csv['county_id'] = [int(cid) for cid in county_csv['county_id']]
county_csv['dem_votes'] = dem_votes_col
county_csv['rep_votes'] = rep_votes_col
county_csv['total_votes'] = [int(tv) for tv in county_csv['total_votes']]

    
county_csv.head(10)
```

```python
state_csv = pd.read_csv("usmap-state.csv")
state_csv.head(10)
```

```python
state_csv = state_csv[state_csv.name.notnull()]
del state_csv['crimerate']
state_total_votes_col = []
state_dem_votes_col = []
state_rep_votes_col = []
total_votes_series = county_csv[['total_votes', 'state_id']]
# print(county_csv.loc[county_csv['state_id'] == 4.0,['state_id','total_votes']].sum(axis=0)[1])
for idx in range(len(state_csv)):
    state_id = state_csv.at[idx, 'state_id']
    # aggregate total votes (population) over state's counties
    state_total_votes = county_csv.loc[county_csv['state_id'] == state_id, ['state_id', 'total_votes']].sum(axis=0)[1]
    state_total_votes_col.append(int(state_total_votes))
    # aggregate dem votes over state's counties
    state_dem_votes = county_csv.loc[county_csv['state_id'] == state_id, ['state_id', 'dem_votes']].sum(axis=0)[1]
    state_dem_votes_col.append(int(state_dem_votes))
    # aggregate rep votes over state's counties
    state_rep_votes = county_csv.loc[county_csv['state_id'] == state_id, ['state_id', 'rep_votes']].sum(axis=0)[1]    
    state_rep_votes_col.append(int(state_rep_votes))
    
state_csv['state_id'] = [int(sid) for sid in state_csv['state_id']]
state_csv['total_votes'] = state_total_votes_col
state_csv['dem_votes'] = state_dem_votes_col
state_csv['rep_votes'] = state_rep_votes_col
#state_csv.rename(columns={'crimerate':'dem_voteshare'}, inplace=True)
state_csv.head(10)
```

```python
state_csv.to_csv("election-states.csv")
county_csv.to_csv("election-counties.csv")
```
